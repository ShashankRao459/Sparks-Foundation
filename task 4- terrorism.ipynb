## Modifying the data

```python

df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'state',
                     'region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed',
                     'nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
                     'weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
```

### keeping only Necessary data

```python
dataT=df[['Year','Month','Day','Country','state','Region','city','latitude',
               'longitude','AttackType','Killed','Wounded',
               'Target','Summary','Group','Target_type','Weapon_type','Motive']]
```

```python
dataT
```

```python

dataT['Wounded'] = dataT['Wounded'].fillna(0)
dataT['Killed'] = dataT['Killed'].fillna(0)

dataT['Casualities'] = dataT['Killed'] + dataT['Wounded']
dataT['Casualities'].sum()
```

## features analysis

```python
plt.figure(figsize=(40,20))
dataT.hist(figsize=(20,20))
```

## Number Of Terrorist Activities Each Year

```python
plt.subplots(figsize=(15,10))
sns.countplot('Year',data = dataT,palette = "cool")
plt.xticks(rotation=90)
plt.title('Number Of Terrorist Activities Each Year')
plt.grid()
plt.show()
```

## Casualities

```python
dat1=dataT[["Year","Killed","Wounded","Casualities"]].groupby("Year").sum()
print(dat1.head(5))

plt.rcParams["figure.figsize"] = (32,20)
dat1.plot(y=["Killed", "Wounded"], kind="bar")

plt.xticks(rotation = 50)
plt.xlabel('Years',fontsize=25)
plt.ylabel('Number of kills and wounded',fontsize=25)
plt.title('KILLED AND WOUNDED', fontsize=35)
plt.show()
```

## Most affected regions

```python
pd.crosstab(dataT.Year, dataT.Region).plot(kind='area',stacked=False,figsize=(20,10))
plt.title('Terrorist Activities By Region In Each Year',fontsize=25)
plt.ylabel('Number of Attacks',fontsize=20)
plt.xlabel("Year",fontsize=20)
plt.show()
```

```python
regT=dataT[["Region","Killed","Wounded","Casualities"]].groupby("Region").sum().sort_values(by="Casualities",ascending=False)
plt.rcParams["figure.figsize"] = (15,6)
regT.plot(y=["Killed", "Wounded"], kind="bar")

plt.xticks(rotation = 45)
plt.xlabel('Region',fontsize=15)
plt.grid(color='gray', linestyle='--', linewidth=0.3)
plt.ylabel('Killed and Wounded',fontsize=15)
plt.title('REGION WISE KILLS AND WOUNDS', fontsize=20)
plt.show()
```

## Most affected Countries

```python
plt.subplots(figsize=(20,10))
sns.barplot(dataT['Country'].value_counts()[:10].index,dataT['Country'].value_counts()[:10].values,palette='cool')
plt.title('Top Countries Affected')
plt.xlabel('Countries')
plt.grid(color='gray', linestyle='--', linewidth=0.3)
plt.ylabel('Count')
plt.show()
```

```python
regT=dataT[["Country","Killed","Wounded","Casualities"]].groupby("Country").sum().sort_values(by="Casualities",ascending=False).head(10)
plt.rcParams["figure.figsize"] = (15,6)
regT.plot(y=["Killed", "Wounded"], kind="bar")

plt.xticks(rotation = 45)
plt.xlabel('country',fontsize=15)
plt.ylabel('Killed and Wounded',fontsize=15)
plt.title('COUNTRY WISE KILLS AND WOUNDS', fontsize=20)
plt.grid(color='gray', linestyle='--', linewidth=0.3)
plt.show()
regT
```

## Most affected States

```python
State_data = dataT['state'].value_counts().nlargest(n=15)
plt.figure(figsize= (18,10))
sns.barplot(x = State_data.index , y = State_data.values)
plt.title('States with most attacks')
plt.xlabel('States')
plt.ylabel('Attacks count')
plt.xticks(rotation = 90)
plt.show()
State_data
```

## Most affected cities

```python
City_data = dataT['city'].value_counts().nlargest(n=15)
plt.figure(figsize= (18,10))
sns.barplot(x = City_data.index , y = City_data.values)
plt.title('Cities with most attacks')
plt.xlabel('Cities')
plt.ylabel('Attacks count')
plt.xticks(rotation = 90)
plt.show()
City_data
```

```python
regT=dataT[["state","Killed","Wounded","Casualities"]].groupby("state").sum().sort_values(by="Casualities",ascending=False).head(10)
plt.rcParams["figure.figsize"] = (15,6)
regT.plot(y=["Killed", "Wounded"], kind="bar")

plt.xticks(rotation = 45)
plt.xlabel('city',fontsize=15)
plt.ylabel('Killed and Wounded',fontsize=15)
plt.title('CITY WISE KILLS AND WOUNDS', fontsize=20)
plt.show()
regT
```

## Attack Type analysis

```python
dataT['AttackType'].value_counts().plot(kind='bar',figsize=(20,10),color=['Indigo','Blue','Green','Yellow','Orange','Red'])
plt.xticks(rotation = 50)
plt.xlabel("AttackType",fontsize=18)
plt.ylabel("Number of attack",fontsize=15)
plt.title("Type of Attack",fontsize=20)
plt.xticks(rotation = 45)
plt.grid(color='gray', linestyle='--', linewidth=0.3)
plt.show()
```

```python
dataT[['AttackType',"Killed","Wounded"]].groupby(["AttackType"],axis=0).sum().sort_values(by="Wounded",ascending=False).plot(kind='bar',figsize=(20,10))
plt.xticks(rotation=50)
plt.title("Number of wounded  ",fontsize=20)
plt.ylabel('Number of people',fontsize=15)
plt.xlabel('Attack type',fontsize=20)
plt.grid(color='gray', linestyle='--', linewidth=0.3)
plt.show()
```

## Target analysis

```python
plt.subplots(figsize=(20,10))
sns.countplot(dataT["Target_type"],order=dataT['Target_type'].value_counts().index,palette="cool",edgecolor=sns.color_palette("mako"));
plt.xticks(rotation=90)
plt.xlabel("Attacktype",fontsize=15)
plt.ylabel("count",fontsize=15)
plt.title("Targets",fontsize=20)
plt.grid(color='gray', linestyle='--', linewidth=0.3)
plt.show()
```

## most active grups

```python
plt.subplots(figsize=(11,5))
sns.barplot(y=dataT['Group'].value_counts()[1:12].index,x=dataT['Group'].value_counts()[1:12].values)
plt.title('Most Active Terrorist Organizations')
plt.grid(color='gray', linestyle='--', linewidth=0.3)
plt.show()
```

```python
gcw=dataT[["Group","Killed","Wounded"]].groupby("Group").sum().sort_values(by="Wounded",ascending=False).drop("Unknown")
gcw.head(10).plot(kind="bar",figsize=(15,8))
plt.title("Casualties from Group Name",fontsize=16)
plt.xlabel("Group Name",fontsize=13)
plt.xticks(fontsize=12)
plt.xticks(rotation=45)
plt.ylabel("Number of Affected Person",fontsize=12)
plt.grid(color='gray', linestyle='--', linewidth=0.3)
plt.show()
gcw.head()
```
